# LINEAR REGRESSION

```python


import pandas as pd
import numpy as np

import matplotlib.pyplot as plt  #visualization library
import seaborn as sns  #visualization library
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,accuracy_score
%matplotlib inline


#linear regression
#attributes-->An attribute is a property of an instance that may be used to determine its classification. 
#In the following dataset, the attributes are the petal and sepal length and width. It is also known as Features.



#Dividing the dataset into 2 components
x = df1.iloc[:, 1:2].values  # we take 2D array so double third bracket
y = df1.iloc[:,3].values


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)

model=LinearRegression() #Fitting Linear Regression to the dataset

model.fit(x_train,y_train)
y_pred=model.predict(x_test)

accuracy=r2_score(y_test,y_pred)
print(r2_score(y_test,y_pred))

plt.scatter(x_train,y_train,color='green')
plt.plot(x_test,y_pred,color='blue')

plt.xlabel("temperature") 
plt.ylabel("precipitation")
plt.title('liner regression')
plt.show()


```

# POLYNOMIAL REGRESSION

```python
#regrassion:- a measure of the relation between the mean value of one variable (e.g. output) and corresponding values 
#of other variables (e.g. time and cost).


import pandas as pd
import numpy as np

import matplotlib.pyplot as plt  #visualization library
import seaborn as sns  #visualization library
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,accuracy_score
%matplotlib inline



#polynomial 
#Regression: if the desired output consists of one or more continuous variables, then the task is called regression. 
#An example of a regression problem would be the prediction of the length of a salmon as a function of its age and weight.



#Dividing the dataset into 2 components

x=df1.iloc[:, 1:2].values # we take 2D array so double third bracket
y=df1.iloc[:, 2].values



model=LinearRegression() #Fitting Linear Regression to the dataset


poly=PolynomialFeatures(degree=21)  # Fitting Polynomial Regression to the dataset
x_poly=poly.fit_transform(x)



model.fit(x_poly,y)   #Fitting the Polynomial Regression model on two components X and y.


y_pred=model.predict(x_poly)

print('Accuracy')
ac2=r2_score(y,y_pred)  # for accuracy
print(r2_score(y,y_pred))

plt.scatter(x, y, color = 'red') 
plt.plot(x,y_pred,color = 'green')

plt.xlabel("MAXIMUM")
plt.ylabel("precipitation")
plt.title('Polynomial Regression') 
plt.show()





  
#plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'red') 

  


```

# DECISION TREE

```python
# Load libraries
import pandas as pd
from sklearn import tree # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
import matplotlib.pyplot as plt

col_names = ['maximum', 'minimum','average','precipitation','snowfall','snow depth','decision taken']
# load dataset
df = pd.read_csv(r'Weather_forecasting.csv',names=col_names, header = 0) # names=col_names, header = 0 write this or error will come

df.head()

#split dataset in features and target variable
feature_cols = ['maximum', 'minimum', 'average', 'precipitation']
X = df[feature_cols] # Features
y = df['decision taken'] # Target variable

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

# Create Decision Tree classifer object
clf = tree.DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

from sklearn.externals.six import StringIO  
from IPython.display import Image 
import pydotplus

dot_data = StringIO()
tree.export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  

graph.write_png('d1.png')
Image(graph.create_png())
plt.show()

```
